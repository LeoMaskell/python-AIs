```python
import numpy as np
import matplotlib.pyplot as plt
import auto_diff as ad
```

```python
def f(a, x, y):
    tmp = a.T@x-y
    return tmp@tmp.T

#3 data points with 2 features
train_data = train.to_numpy()
test_data = test.to_numpy()

newx = test_data[:,0:8].T

y = train_data[:,13]
x = train_data[:,0:8].T
aval = np.zeros((8, 1))
#y = np.array([[1,4,3]]) #all the predictions
# x = np.array([[2,1,67], [1,45,33]]) #all the features - independent var
#aval = np.array([[2], [4]]) #parameters - stuff we change
#print(aval.T@x-y)
#print(f(aval, x, y))
#print(y)
```

```python
#probability and prediction
Delta = 0.000000001 #change for different k num.


for i in range(1000):
    print(f(aval,x,y))

    with ad.AutoDiff(aval) as a:
        ignoreme, j = ad.get_value_and_jacobian(f(a, x, y))
    aval = aval-Delta*j.T

# the numbers for confidence, is the total of how many it gets wrong

#newx = np.array([[5],[4]])
print(aval.T@newx) #the pred. for y
```
